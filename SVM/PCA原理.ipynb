```python
# PCA降维，通过矩阵运算

import numpy as np
import sklearn.datasets as datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
iris=datasets.load_iris()
X=iris["data"]
y=iris["target"]
```

```python
# whiten  是对数据进行了Standard 归一化
s=StandardScaler()
pca=PCA(n_components=0.95,whiten=False)
X_pca=pca.fit_transform(X)
s.fit_transform(X_pca)
```

```python
pca=PCA(n_components=0.95,whiten=True)
X_pca=pca.fit_transform(X)
X_pca
```

```python
# 1. 去中心化
# 2. 协方差
# 3. 特征值和特征向量计算
# 4. 从特征值中选取白分比大于95%的特征所对应的特征向量
# 5. 使用选取数来的特征向量进行矩阵运算
```

```python
# 1. 去中心化
A=X-X.mean(axis=0)
A
```

```python
# 2. 协方差
V=np.cov(A,rowvar=False)
V
```

```python
# 3. 特征值和特征向量计算

T,TV=np.linalg.eig(V)
print(T,TV)
```

```python
#   选取前两个的特性向量
#  4. 从特征值中选取白分比大于95%的特征所对应的特征向量
P=TV[:,[0,1]]
# 5. 使用选取数来的特征向量进行矩阵运算
s=StandardScaler()
s.fit_transform(X.dot(P))
```
