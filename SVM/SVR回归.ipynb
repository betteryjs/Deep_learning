```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.svm import SVR
import seaborn as sns
sns.set_style(style="darkgrid")
```

```python
X=np.linspace(0,3*np.pi,50).reshape(-1,1)
y=np.sin(X)
```

```python
plt.scatter(x=X,y=y)
```

```python
svr_linear=SVR(kernel="linear")
svr_rbf=SVR(kernel="rbf")

svr_poly=SVR(kernel="poly") # 多项式

svr_linear.fit(X,y)
svr_rbf.fit(X,y)

svr_poly.fit(X,y)
```

```python
X_test=np.linspace(0,3*np.pi,128).reshape(-1,1)
```

```python
y1=svr_linear.predict(X_test)
y2=svr_rbf.predict(X_test)
y3=svr_poly.predict(X_test)
```

```python

plt.scatter(x=X,y=y)
plt.plot(X_test,y1)
plt.plot(X_test,y2)
plt.plot(X_test,y3)
plt.legend(["linear","rbf","poly"])
```
