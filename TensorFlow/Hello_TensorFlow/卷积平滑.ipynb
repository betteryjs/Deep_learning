```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
tf.compat.v1.disable_eager_execution()
```

```python
moon=plt.imread("moonlanding.png")
moon.shape
```

```python
plt.figure(figsize=(12,9))
plt.imshow(moon,cmap="gray")
```

```python

filters_=np.full(shape=[3,3],fill_value=1/9).astype(np.float32).reshape(3,3,1,1)

conv=tf.nn.conv2d(input=moon.reshape(1,474,630,1),filters=filters_,strides=[1,1,1,1],padding="SAME")
with tf.compat.v1.Session() as sess:
    res=sess.run(conv)
    plt.figure(figsize=(12,9))
    plt.imshow(res.reshape(474,630),cmap="gray")
```

```python

filters_=np.array([[1/16,2/16,1/16],[2/16,4/16,2/16],[1/16,2/16,1/16]]).astype(np.float32).reshape(3,3,1,1)

conv=tf.nn.conv2d(input=moon.reshape(1,474,630,1),filters=filters_,strides=[1,1,1,1],padding="SAME")
with tf.compat.v1.Session() as sess:
    res=sess.run(conv)
    plt.figure(figsize=(12,9))
    plt.imshow(res.reshape(474,630),cmap="gray")
```
