```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
tf.compat.v1.disable_eager_execution()
```

```python
plt.figure(figsize=(12,9))
image=plt.imread("./欧式.jpg")
plt.imshow(image)
```

```python
image.shape
```

```python
# filters : [filter_height * filter_width * in_channels, output_channels] 

filters_=np.full(shape=[3,3,3],fill_value=1/27).reshape(3,3,3,1)
conv=tf.nn.conv2d(
    input=image.reshape(1,582,1024,3).astype(np.float32),
    filters=filters_,
    strides=[1,1,1,1],
    padding="SAME"
)
with tf.compat.v1.Session() as sess:
    ret=sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    plt.imshow(ret.reshape(582,1024),cmap="gray")
    
```

```python
# filters : [filter_height * filter_width * in_channels, output_channels] 

filters_=np.random.normal(loc=1/27,size=(3,3,3,3))
conv=tf.nn.conv2d(
    input=image.reshape(1,582,1024,3).astype(np.float32),
    filters=filters_,
    strides=[1,1,1,1],
    padding="SAME"
)
with tf.compat.v1.Session() as sess:
    ret=sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    plt.imshow(ret.reshape(582,1024,3),cmap="gray")
```

```python
image=np.transpose(image,axes=[2,0,1])
image.shape
```

```python
# filters : [filter_height * filter_width * in_channels, output_channels] 

filters_=np.full(shape=[3,3],fill_value=1/9).reshape(3,3,1,1)
conv=tf.nn.conv2d(
    input=image.reshape(3,582,1024,1).astype(np.float32),
    filters=filters_,
    strides=[1,1,1,1],
    padding="SAME"
)
with tf.compat.v1.Session() as sess:
    ret=sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    
    plt.imshow(ret.reshape(3,582,1024).transpose([1,2,0]).astype(np.uint8))
```

```python
bridge=plt.imread("./欧式.jpg")[380:520,180:780]
plt.imshow(bridge)
```

```python
# filters : [filter_height * filter_width * in_channels, output_channels] 
number=8
filters_=np.full(shape=[number,number],fill_value=1/(number**2)).reshape(number,number,1,1)
conv=tf.nn.conv2d(
    input=image.reshape(3,582,1024,1).astype(np.float32),
    filters=filters_,
    strides=[1,1,1,1],
    padding="SAME"
)
with tf.compat.v1.Session() as sess:
    ret=sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    img=ret.reshape(3,582,1024).transpose([1,2,0]).astype(np.uint8)
    img[380:520,180:780]=bridge
    plt.imshow(img)
```
