```python
import numpy as np
import sklearn.datasets as datasets
import pandas as pd
from sklearn.feature_selection import VarianceThreshold # 根据方差来获取

from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, chi2
```

```python
iris = datasets.load_iris()
X = iris['data']
X
```

```python
X.var(axis=0)
```

```python
# Feature selector that removes all low-variance features. 把方差低的(没有波动结果稳定)的筛选出来
# 特征如何没有波动不足以作为训练的数据

v=VarianceThreshold(threshold=3)
```

```python
X2=v.fit_transform(X)
X2
```

```python
X2.var(axis=0)
```

```python
# 计算相关性系数
np.corrcoef(X)
```
