```python
# 如果有截距，求解时，需要梯度下降法求解w和b
```

```python
# 什么是梯度？？？
# 梯度就是导数
# 梯度下降是求解最大值，最小值最快的速度
# 导数是函数变化率最快的方向
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f=lambda x:(x-4.3)**2+7.5*x+6
#  0.55
g=lambda x:2*x-1.1
x=np.linspace(-5,9,100)
plt.plot(x,f(x))
```

```python
# 设置起始值的数值
result=[]
v_min=np.random.randint(-10,10,size=(1))[0]
# 该值记录梯度下降过程中，上一步的值
result.append(v_min)
v_min_last=v_min+1

# 精度的退出条件 
precision=0.0001

# 再一个退出条件 循环次数
max_time=3000

# 布幅
step=0.01

count=0
print("------------------随机生成的最小值  : ",v_min)
while True:
    if np.abs(v_min - v_min_last) < precision:
        break
    if count > max_time:
        break
    v_min_last=v_min
    v_min=v_min-g(v_min)*step
    result.append(v_min)
    print("------------------梯度下降更新的最小值  : ",v_min)
    count=count+1
```

```python
import seaborn as sns
sns.set(style="darkgrid")
plt.figure(figsize=(12,9))
x=np.linspace(-3,5,100)
sns.lineplot(x,f(x))

# 绘制最小值的点

plt.scatter(
    x=result,
    y=f(np.array(result)),
    color="r"
)

# plt.scatter(result,)
```
