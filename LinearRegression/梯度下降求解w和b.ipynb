### 如果要求斜率 LinearRegression 也使用的是梯度下降算法


![](1.png)

```python

# 类的目的是求解斜率和截距

class Liner_model(object):
    def __init__(self):
        self.w=np.random.randn(1)[0]
        self.b=np.random.randn(1)[0]
        self.precisoin=0.00001
        self.max_count=3000
        print("-------------------",self.w,self,b)

        
        
    def model(self,x):
        """
        model 是方程 f(x)=wX+b
        """
        return self.w*x+self.b
    
    def loss(self,x,y):
        """
        线性问题 原理都是最小2乘法
        """
        cost=(y-self.model(x))**2
        
        # 求偏导数
        g_w=2*(y-self.model(x))*(-x)
        g_b=2*(y-self.model(x))*(-1)
        return g_w,g_b
    
    
    def gradient_descend(self,g_w,g_b,step=0.01):
        # 更新新的斜率和截距
        self.w=self.w-g_w*step
        self.b=self.b-g_b*step
        print("-------------------",self.w,self,b)
    
    def fit(self,X,y):
        
        w_last=self.w+1
        b_last=self.b+1
        count=0
        while True:
            if (np.abs(self.w-w_last)< self.precisoin) and (np.abs(self.b-b_last)< self.precisoin):
                break
            if count> self.max_count:
                break
            # 更新斜率和截距
            g_w=0
            g_b=0
            size=X.shape[0]
            for xi,yi in zip(X,y):
                g_w+=self.loss(xi,yi)[0]/size
                g_b+=self.loss(xi,yi)[1]/size
            
            self.gradient_descend(g_w,g_b)
            count=count+1
            
            
    def coef_(self):
        return self.w
    
    def intercept_(self):
        return self.b
        
        
        
        
        
```

```python
lm=Liner_model()
```

```python
lm.fit(X,y)
```

```python
lm.coef_()
```

```python
lm.intercept_()
```
